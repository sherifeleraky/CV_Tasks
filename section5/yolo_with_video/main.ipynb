```python
from PIL import Image

from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolo11m.pt")

path = "person-bicycle-car-detection.mp4"
results = model(source=path,conf=0.4)
```

```python
frames = []


for frame in results:
    frames.append(frame.plot())
```

```python
import cv2


first_image = frames[0]


frame_size = (first_image.shape[1], first_image.shape[0]) 


output_filename = 'output_video.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID') 
fps = 24  

out = cv2.VideoWriter(output_filename, fourcc, fps, frame_size)

for frame in frames:
    out.write(frame)
   
out.release()
print("Video created successfully.")
```

```python
import subprocess
import os

def convert_avi_to_mp4(input_file, output_file):
    # Check if input file exists
    if not os.path.exists(input_file):
        print(f"Input file '{input_file}' not found.")
        return
    
    # Remove existing output file to avoid conflicts
    if os.path.exists(output_file):
        os.remove(output_file)
    
    # Run ffmpeg command
    try:
        command = ['ffmpeg', '-i', input_file, output_file]
        subprocess.run(command, check=True)
        print("Video converted successfully.")
    except subprocess.CalledProcessError as e:
        print(f"Error during conversion: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Input and output file paths
input_file = 'output_video.avi'
output_file = 'output_video.mp4'

convert_avi_to_mp4(input_file, output_file)
```
